```python
from __future__ import print_function

import histomicstk as htk

import numpy as np
import scipy as sp

import skimage.io
import skimage.measure
import skimage.color

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline

#Some nice default configuration for plots
plt.rcParams['figure.figsize'] = 15, 15
plt.rcParams['image.cmap'] = 'gray'
titlesize = 24
```

```python
inputImageFile =('/media/raj/Raj1_5/10ktiles/train/gbm/TCGA-02-0001-01Z-00-DX2_20x_13536_6944_256x256.png')  # H&E.png

imInput = skimage.io.imread(inputImageFile)[:, :, :3]

plt.imshow(imInput)
_ = plt.title('Input Image', fontsize=16)
```

```python
# create stain to color map
stain_color_map = htk.preprocessing.color_deconvolution.stain_color_map
print('stain_color_map:', stain_color_map, sep='\n')

# specify stains of input image
stains = ['hematoxylin',  # nuclei stain
          'eosin',        # cytoplasm stain
          'null']         # set to null if input contains only two stains

# create stain matrix
W = np.array([stain_color_map[st] for st in stains]).T

# perform standard color deconvolution
imDeconvolved = htk.preprocessing.color_deconvolution.color_deconvolution(imInput, W)

# Display results
for i in 0, 1:
    plt.figure()
    plt.imshow(imDeconvolved.Stains[:, :, i])
    _ = plt.title(stains[i], fontsize=titlesize)
```

```python
# Calculating hematoxylin average
h_avg = (65+70+29)/3
```

```python
# Calculating Eosin Avaerage
e_avg=(7+99+11)/3
```

```python
import cv2
img = cv2.imread('/media/raj/Raj1_5/10ktiles/train/gbm/TCGA-02-0001-01Z-00-DX2_20x_13536_6944_256x256.png')
```

```python
#splitting the image into BGR channels
(channel_b, channel_g, channel_r) = cv2.split(img)
```

```python
#Calculating Image average
channel_out = []
for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            avg = (img[i,j,0]+img[i,j,1]+img[i,j,2])/3
            channel_out.append(avg)
```

```python
len(channel_out)
```

```python
channel_purple =[]
channel_pink =[]
channel_white =[]
for i in range(len(channel_out)):
   x=channel_out[i]
   if x<= h_avg: channel_purple.append(x)
   if x>h_avg and x<=e_avg : channel_pink.append(x)
   if x>h_avg and x>e_avg: channel_white.append(x)
```
